```python
text = ""

with open("/Users/minjikim/Downloads/사고뭉치.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        text = re.sub(r'[0-9]+', '', text)
        text = re.sub('[-=.#/?:$}]','',text)
        text += line.replace('요정민','').replace('애뤼','').replace('김민지','').replace('만경','').replace('User','').replace('Message','').replace('Date','').replace('ㅋ','').replace('ㅎ','').replace('ㅠ','').replace('ㅜ','').replace('사진','').replace('이모티콘','').replace('샵검색','').replace('나','').replace('근데','').replace('아','').replace('진짜','').replace('아니','').replace('너무','').replace('내가','').replace('난','').replace('니','').replace('동영상','').replace('도','').replace('다','').replace('유리','')
print(text)
```

```python
text=text.replace('\t','')
print(text)
```

```python
for i in text:
    if len(i)>2:
        text+=i
print(text)
```
